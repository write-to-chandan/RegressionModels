```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
X = df[['fixed', 'volatile', 'citric', 'sugar', 'chlorides', 'free', 'total',
       'density', 'pH', 'sulphates', 'alcohol']]
y = df['quality']
```

## Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score,mean_squared_error
import numpy as np
from sklearn import metrics

df = pd.read_csv("/Users/chandankumar/Documents/Datasets/WineData3a.csv", header=0)
df.head()
```

```python
X = df[['fixed', 'volatile', 'citric', 'sugar', 'chlorides', 'free', 'total',
       'density', 'pH', 'sulphates', 'alcohol']]
y = df['quality']
```

```python
lst = [0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 0.85, 0.9, 0.95]
for ls in lst:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=(1-ls), random_state=0)
    regr_1 = DecisionTreeRegressor(max_depth=10)
    regr_1.fit(X_train, y_train)
    y_predict = regr_1.predict(X_test)
    mse = mean_squared_error(y_test, y_predict)
    rmse = np.sqrt(mse)
    print("\n---------------------------------")
    print("Test Size: ", (1-ls))
    print("-----------------------------------")
    print("Root Mean Square Error: ", rmse)
    df2 =pd.DataFrame({'Actual':y_test, 'Predicted':y_predict})
    df2['diff'] = np.abs(pd.eval('(y_test-y_predict)/y_test'))
    df2['acc'] = 1 - df2['diff']
    accuracy = df2['acc'].mean()
    print("Accuracy is: ", accuracy)
    print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_predict))
    print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_predict))
    print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_predict)))
    print("Variance :", np.var(pd.eval('y_test-y_predict')))
    print("R Squred :", r2_score(y_test,y_predict))
```

## Gradient Boosted Decision Tree Regressor

```python
from sklearn.datasets import make_regression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split


import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.metrics import mean_squared_error

from sklearn import ensemble
from sklearn.metrics import mean_squared_error, r2_score


df = pd.read_csv("/Users/chandankumar/Documents/Datasets/WineData3a.csv", header=0)
df.head()
```

```python
X = df[['fixed', 'volatile', 'citric', 'sugar', 'chlorides', 'free', 'total',
       'density', 'pH', 'sulphates', 'alcohol']]
y = df['quality']
```

```python
lst = [0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 0.85, 0.9, 0.95]
for ls in lst:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=(1-ls), random_state=234, shuffle=True)
    params = {'n_estimators': 500, 'max_depth': 4, 'min_samples_split': 2, 'learning_rate': 0.01, 'loss': 'ls'}
    model = ensemble.GradientBoostingRegressor(**params)
    model.fit(X_train, y_train)
    model_score = model.score(X_train,y_train)
    print("\n---------------------------------")
    print("Test Size: ", (1-ls))
    print("-----------------------------------")
    print('R Squared : ', model_score)
    y_predicted = model.predict(X_test)
    # The mean squared error
    print("Mean squared error: %.2f"% mean_squared_error(y_test, y_predicted))
    print("Root Mean squared error: ", np.sqrt(mean_squared_error(y_test, y_predicted)))
    # Explained variance score: 1 is perfect prediction
    print('Test Variance score: %.2f' % r2_score(y_test, y_predicted))
    df3 =pd.DataFrame({'Actual_1':y_test, 'Predicted_1':y_predicted})
    df3['diff_1'] = np.abs(pd.eval('(y_test-y_predicted)/y_test'))
    df3['acc_1'] = 1 - df3['diff_1']
    accuracy_1 = df3['acc_1'].mean()
    print("Accuracy is: ", accuracy_1)
    
```

## Random Forest Regressor

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.metrics import mean_squared_error

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

df = pd.read_csv("/Users/chandankumar/Documents/Datasets/WineData3a.csv", header=0)
df.head()
```

```python
X = df[['fixed', 'volatile', 'citric', 'sugar', 'chlorides', 'free', 'total',
       'density', 'pH', 'sulphates', 'alcohol']]
y = df['quality']
```

```python
lst = [0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 0.85, 0.9, 0.95]
for ls in lst:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=(1-ls), random_state=234, shuffle=True)
    # params = {'n_estimators': 500, 'max_depth': 4, 'min_samples_split': 2, 'learning_rate': 0.01, 'loss': 'ls'}
    n_estimators=100
    model = RandomForestRegressor(random_state=0, n_estimators=n_estimators)
    model.fit(X_train, y_train)
    model_score = model.score(X_train,y_train)
    print("\n---------------------------------")
    print("Test Size: ", (1-ls))
    print("-----------------------------------")
    # print('coefficient of determination R^2 of the prediction.: ',model_score)
    print('R Squared : ', model_score)
    y_predicted = model.predict(X_test)
    # The mean squared error
    print("Mean squared error: %.2f"% mean_squared_error(y_test, y_predicted))
    print("Root Mean squared error: ", np.sqrt(mean_squared_error(y_test, y_predicted)))
    # Explained variance score: 1 is perfect prediction
    print('Test Variance score: %.2f' % r2_score(y_test, y_predicted))
    df4 =pd.DataFrame({'Actual_2':y_test, 'Predicted_2':y_predicted})
    df4['diff_2'] = np.abs(pd.eval('(y_test-y_predicted)/y_test'))
    df4['acc_2'] = 1 - df4['diff_2']
    accuracy_2 = df4['acc_2'].mean()
    print("Accuracy is: ", accuracy_2)
    
```
