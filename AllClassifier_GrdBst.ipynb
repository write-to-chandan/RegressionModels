## AllClassifier Gradient Boosteed Classifier

```python
from sklearn.datasets import make_classification
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn.preprocessing import OneHotEncoder
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn import datasets
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error


def condition(r):
    if (0 <= r <= 5):
        label = 4
    else:
        label = 7
    return label
```

```python
df = pd.read_csv("/Users/chandankumar/Documents/Datasets/WineData3a.csv", header=0)
df.head()
```

```python
df["quality_class"] = df['quality'].apply(lambda x: condition(x))
df.head()
```

```python
X = df[['fixed', 'volatile', 'citric', 'sugar', 'chlorides', 'free', 'total',
       'density', 'pH', 'sulphates', 'alcohol']]
y = df['quality_class']
```

```python
lst = [0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 0.85, 0.9, 0.95]
for ls in lst:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=(1-ls), random_state=0)
    gbClassifier = GradientBoostingClassifier(random_state=0)
    gbClassifier.fit(X_train, y_train)
    predictions = gbClassifier.predict(X_test)
    score = gbClassifier.score(X_test, y_test)
    # Use score method to get accuracy of model
    print("\n---------------------------------")
    print("Test Size: ", round(1-ls,2))
    print("-----------------------------------")
    print("Accuracy: ", score)
    # print(score)
    cm = metrics.confusion_matrix(y_test, predictions)
    tn, fp, fn, tp = metrics.confusion_matrix(y_test, predictions).ravel()
    print("Confusion Matrix: ")
    print(cm)
    sensitivity_or_recall = tp/(tp+fn)
    specificity = tn/(tn+fp)
    precison = tp/(tp+fp)
    F1Score = (2*tp)/((2*tp)+fp+fn)
    print("Recall: ", sensitivity_or_recall)
    print("Specificity: ", specificity)
    print("Precision: ", precison)
    print("F1 - Score: ", F1Score )
    r2 = r2_score(y_test, predictions)
    print("R Squared: ", r2)
    # print(r2)
    mse = mean_squared_error(y_test, predictions)
    print("Mean Squared Error", np.sqrt(mse))
```
